```python
from transformers import pipeline
import sklearn_json as skjson
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("final_to_gpt.csv")
```

```python
data
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
reviews = data['review_lemmatized']
```

```python
senti_scores = []
for i in range(len(reviews)):
    if i == 33:
        continue
    score = sentiment_pipeline(reviews[i])
    senti_scores.append(score)
```

```python
senti_scores[0]
```

```python
senti_scores.insert(33,{'label': 'NEGATIVE', 'score': 0.58})
```

```python
data['overall_sentiment_score'] = senti_scores
```

```python
data.to_csv("final_to_gpt.csv")
```
